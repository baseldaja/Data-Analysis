```python
missing_exp=apps["relevent_experience"].isnull()
apps[missing_exp]
```

<span style="font-size: 20px; color: lightblue;">Result: Missing enrolled university filled with "no_enrollment"</span>

```python
missing_edulvl=apps["education_level"].isnull()
apps[missing_edulvl]
```

<span style="font-size: 20px; color: lightblue;">Result: 460 Rows with missing education level<br>so there is three options:<ol><li>Fill with "Not provided"</li><li>Drop Rows</li><li>Fill with the mode</li></ol></span>

```python
clean_apps['education_level']=apps['education_level'].fillna('Not provided') # I choose to fill with "Not provided"
```

<span style="font-size: 20px; color: lightblue;">Result: Missing education level filled with "Not provided"</span>

```python
missing_major=apps["major_discipline"].isnull()
apps[missing_major]
```

<span style="font-size: 20px; color: lightblue;">Result: 2813 Missing Rows with missing major discipline</span>

```python
clean_apps["major_discipline"].unique() #shows the unique values in specific feature 
clean_apps['major_discipline']=apps['major_discipline'].fillna('Other') # I choose to fill with "Other"
```

<span style="font-size: 20px; color: lightblue;">Result: Missing major discipline filled with "No Major"</span>

```python
missing_yoexp=apps["experience"].isnull()
apps[missing_yoexp]
```

<span style="font-size: 20px; color: lightblue;">Result: 65 Rows with missing years of experience<br>**I could drop rows with no experience<br>**But i'll fill the missings with 0 years</span>

```python
clean_apps['experience']=apps['experience'].fillna(0) # I choose to fill with "0"
```

<span style="font-size: 20px; color: lightblue;">Result: Missing years of experience filled with "0"</span>

```python
missing_compsz=apps["company_size"].isnull()
apps[missing_compsz]
```

<span style="font-size: 20px; color: lightblue;">Result: 5938 Rows with missing company size<br>**I will fill with 0 and we can make the drop decision later on.</span>

```python
clean_apps['company_size']=apps['company_size'].fillna(0) # I choose to fill with 0
```

<span style="font-size: 20px; color: lightblue;">Result: Missing company size filled with 0</span>
