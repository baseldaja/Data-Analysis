```python
#Dealing with missing values by filling with default value (might be the most common and avg used GPU)
df["GPU"]=df["GPU"].fillna("RTX 2060")
df["Final Price"]+=300 #300 is the price of RTX 2060 GPU
```

```python
#Dealing with Duplicates in specific column (in this case i applied on GPU column, making the GPU as a unique value)
laptop_clean=df.drop_duplicates(subset=["GPU"])
print((df['GPU'].unique()==laptop_clean['GPU']).tolist()) #comparison for assurance
```

```python
df['RAM'].value_counts() #showing the count of values in RAM column
```

```python
num_data = df.select_dtypes(include=['float', 'int'])  # Select numerical columns
cat_data = df.select_dtypes(include=['object', 'category'])  # Select categorical columns
```


<span style="font-size: 30px;color:cornflowerblue">Figures And Insights:</span>

```python
#This figure shows the prices of the laptop of a specific GPU
laptop_clean.plot(x='GPU',y="Final Price" ,kind='bar')
```

```python
#This figure shows the price range for each RTX gpu
plt.figure(figsize=(10, 5))
#Filter RTX GPUs
rtx_data = df[df['GPU'].str.startswith('RTX')]

#Group by GPU and calculate min and max prices
rtx_agg = rtx_data.groupby('GPU')['Final Price'].agg(['min', 'max'])

#Reset the index to keep 'GPU' as a column
rtx_agg = rtx_agg.reset_index()


plt.scatter(rtx_agg['GPU'], rtx_agg['min'], color='blue', label='Min Price')
plt.scatter(rtx_agg['GPU'], rtx_agg['max'], color='red', label='Max Price')

#Annotate min and max prices
for i, row in rtx_agg.iterrows():
    plt.text(row['GPU'], row['min'], f"${int(row['min'])}", fontsize=8, ha='center', va='bottom')
    plt.text(row['GPU'], row['max'], f"${int(row['max'])}", fontsize=8, ha='center', va='bottom')
    
# Draw dashed lines between min and max points
    plt.plot([row['GPU'], row['GPU']], [row['min'], row['max']], color='gray', linestyle='--')

plt.xlabel('GPU')
plt.ylabel('Price')
plt.title('Scatter Plot with Min and Max Prices of laptops that had RTX GPUs')
plt.xticks(rotation=70)
plt.legend()
plt.show()
```

```python
#histogram for gpu (rtx 2060 removed because its hight frequency)
gpus_no_rtx2060 = df[df['GPU'] != 'RTX 2060']
fig, axs = plt.subplots(1, 2, figsize=(20, 5)) 


axs[0].hist(df['GPU'],bins=50,edgecolor='black')
axs[0].set_xticks(axs[0].get_xticks())
axs[0].set_xlabel('GPU',color="Blue")
axs[0].set_ylabel('Frequency',color="Blue")
axs[0].tick_params(axis='x', rotation=80)
axs[0].set_title("Histogram for GPUs with RTX 2060")

axs[1].hist(gpus_no_rtx2060['GPU'],bins=50,edgecolor='black')
axs[1].set_xlabel('GPU',color="Blue")
axs[1].set_ylabel('Frequency',color="Blue")
axs[1].set_xticks(axs[1].get_xticks())
axs[1].tick_params(axis='x', rotation=80)
axs[1].set_title("Histogram for GPUs without RTX 2060")

plt.tight_layout()

# Show the plot
plt.show()
```

```python
#Seaborn boxplot on RAM and PRICE
sns.boxplot(x=num_data['RAM'],y=num_data['Final Price'])
plt.show()
```

```python
#Distribution of GPUs
gpu_counts = df['GPU'].value_counts()
plt.figure(figsize=(10, 6))
gpu_counts.plot(kind='bar')
plt.xlabel('GPU')
plt.ylabel('Count')
plt.title('Distribution of GPUs')
plt.xticks(rotation=90)  
plt.tight_layout()
plt.show()
```

```python
#Outliers Example
# Calculate quartiles and IQR
Q1 = df['RAM'].quantile(0.25)
Q3 = df['RAM'].quantile(0.75)
IQR = Q3 - Q1

# Calculate bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# finding outliers
outliers = df[(df['RAM'] < lower_bound) | (df['RAM'] > upper_bound)]

#before removing outliers
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(x=df['RAM'],y=df['RAM'])
plt.title("Normal Scatter Plot For RAMs") 
#after removing outliers
df_no_outliers = df[~((df['RAM'] < lower_bound) | (df['RAM'] > upper_bound*2))] #multiplication by 2 on the upperbound to understand the acceptable outliers
plt.subplot(1, 2, 2)
plt.scatter(x=df_no_outliers['RAM'],y=df_no_outliers['RAM'])
plt.title("Scatter Plot After Removing Outliers From RAMs")

plt.show()
```

```python
#Show GPUs frequency percentage
gpufreq=cat_data['GPU'].value_counts(normalize=True)
threshold=0.01
gpus_to_remove = gpufreq[gpufreq < threshold].index
filtered_cat = cat_data[~cat_data["GPU"].isin(gpus_to_remove)]


plt.figure(figsize=(8,8))
plt.pie(filtered_cat["GPU"].value_counts(), labels=filtered_cat["GPU"].unique(), autopct='%1.1f%%', startangle=140)

plt.title('Laptop GPUs frequency distribution')
plt.show()
```
